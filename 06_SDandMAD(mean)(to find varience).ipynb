```python
import pandas as pd
import numpy as np

# Data with less deviation
data1 = {
    'name': ['a', 'b', 'c', 'd', 'e', 'f'],
    'score': [75, 72, 68, 65, 67, 75]  
}

# Data with higher deviation
data2 = {
    'name': ['a', 'b', 'c', 'd', 'e', 'f'],
    'score': [83, 70, 70, 63, 70, 70]  
}

# Create DataFrames
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Print DataFrames
print(df1)
print(df2)

# Calculate mean of scores
avg1 = df1['score'].mean()
avg2 = df2['score'].mean()

# Add a new column for absolute deviation from the mean
df1['score-avg'] = abs(df1['score'] - avg1)
print(df1)

# Calculate Mean Absolute Deviation (MAD)
MAD1 = df1['score-avg'].mean()
print('MAD1:', MAD1)

# Add a new column for squared deviations
df1['sqr'] = df1['score-avg']**2
print(df1)

# Calculate Standard Deviation (SD)
sd1 = np.sqrt(df1['sqr'].mean())
print('SD1:', sd1)

# Repeat the same process for the second dataset
df2['score-avg'] = abs(df2['score'] - avg2)
print(df2)

MAD2 = df2['score-avg'].mean()
print('MAD2:', MAD2)

df2['sqr'] = df2['score-avg']**2
print(df2)

sd2 = np.sqrt(df2['sqr'].mean())
print('SD2:', sd2)
```

```python
import pandas as pd
import numpy as np

# Data with less deviation
data1 = {
    'name': ['a', 'b', 'c', 'd', 'e', 'f'],
    'score': [75, 72, 68, 65, 67, 75]  
}

# Data with higher deviation
data2 = {
    'name': ['a', 'b', 'c', 'd', 'e', 'f'],
    'score': [83, 70, 70, 63, 70, 70]  
}

# Create DataFrames
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Print DataFrames
print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

# Calculate mean of scores
avg1 = df1['score'].mean()
avg2 = df2['score'].mean()

# Add a new column for absolute deviation from the mean
df1['score-avg'] = abs(df1['score'] - avg1)
print("\nDataFrame 1 with absolute deviation from mean:")
print(df1)

# Calculate Mean Absolute Deviation (MAD)
MAD1 = df1['score-avg'].mean()
print('MAD1 (Mean Absolute Deviation):', MAD1)

# Add a new column for squared deviations
df1['sqr'] = df1['score-avg']**2
print("\nDataFrame 1 with squared deviations:")
print(df1)

# Calculate Standard Deviation (SD)
sd1 = np.sqrt(df1['sqr'].mean())
print('SD1 (Standard Deviation):', sd1)

# Repeat the same process for the second dataset
df2['score-avg'] = abs(df2['score'] - avg2)
print("\nDataFrame 2 with absolute deviation from mean:")
print(df2)

MAD2 = df2['score-avg'].mean()
print('MAD2 (Mean Absolute Deviation):', MAD2)

df2['sqr'] = df2['score-avg']**2
print("\nDataFrame 2 with squared deviations:")
print(df2)

sd2 = np.sqrt(df2['sqr'].mean())
print('SD2 (Standard Deviation):', sd2)

# Calculate Median Absolute Deviation (Median AD)
# For DataFrame 1
median1 = df1['score'].median()
df1['score-median'] = abs(df1['score'] - median1)
MAD_median1 = df1['score-median'].median()
print("\nDataFrame 1 with absolute deviation from median:")
print(df1)
print('MAD_median1 (Median Absolute Deviation):', MAD_median1)

# For DataFrame 2
median2 = df2['score'].median()
df2['score-median'] = abs(df2['score'] - median2)
MAD_median2 = df2['score-median'].median()
print("\nDataFrame 2 with absolute deviation from median:")
print(df2)
print('MAD_median2 (Median Absolute Deviation):', MAD_median2)
```
